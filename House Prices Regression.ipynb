**<h1>Ames Housing Prices Regression</h1>**


## Problem Understanding

**Context**

Ames is a small city in the state of Iowa, United States. According to the 2020 census, Ames had a population of 66,427. And regarding the living condition of the city, Ames has been listed at No. 15 on the year 2020 Top 100 Best Places to Live, according to livability.com.

The trainset contains 1,460 houses sold in Ames. With this past data, we can train a prediction model that can help people in Ames set the appropriate sale price for future listings.

**Problem Statement**



The increase in the number of homeowners (supply), and the number of residents (demand) will affect the price curve. Determining the right price is essential to remain competitive in selling the houses. As a higher price is not desired by the buyers and a lower price is not desired by the homeowners.

**Goals**

Predict the appropriate price for the houses in Ames for future listings.

**Analytic Approach**

Analyze the pattern in the housing data and build a regression model to get the appropriate prices.

**Metric Evaluation**

The model will be evaluated using Root Mean Square Error (RMSE). It measures the residual error which has the same unit size as the response variable.

```python
# create class for dropping columns
class ColumnDropTransformer():
    def __init__(self, columns):
        self.columns=columns

    def drop(self, df):
        if isinstance(self.columns, list):
            for column in self.columns:
                try:
                    df.drop(column, axis=1, inplace=True)
                except:
                    pass
            return df
        else:
            try:
                df.drop(self.columns, axis=1, inplace=True)
                return df
            except:
                return df

    __call__ = drop
```

## Outliers

```python
# show outliers
def find_anomalies(feature):

    data = df[feature]
    min = data.min()
    max = data.max()

    q1 = data.quantile(0.25)
    q2 = data.quantile(0.50)
    q3 = data.quantile(0.75)
    iqr = q3 - q1

    limit = iqr * 1.5
    lower_limit = q1 - limit
    upper_limit = q3 + limit

    lower_outliers = len(df[df[feature] < lower_limit])
    upper_outliers = len(df[df[feature] > upper_limit])

    result = {'feature': feature,
            'iqr': iqr,
            'min': min,
            'lower_limit': lower_limit,
            '25%': q1,
            '50%': q2,
            '75%': q3,
            'upper_limit': upper_limit,
            'max': max,
            'lower_outliers': lower_outliers,
            'upper_outliers': upper_outliers}

    return(result)
```

## Model Benchmarking

```python
# Defining Benchmark Models

dt = DecisionTreeRegressor(random_state=42)
rf = RandomForestRegressor(random_state=42)
ada = AdaBoostRegressor(dt, random_state=42)
gbr = GradientBoostingRegressor(random_state=42)
xgb = XGBRegressor(random_state=42, verbosity=0)
lgbm = LGBMRegressor(random_state=42)
```

```python
# Cross-validation on benchmark models

scaler = StandardScaler()
models = [dt, rf, ada, gbr, xgb, lgbm]
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
score=[]
mean_values=[]
std_values=[]

for i in models:
   
    estimator = Pipeline([
                          ('preprocessing', transformer),
                          ('scaler', scaler),
                          ('model', i)
    ])

    model_cv = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=kfold, 
        scoring='neg_root_mean_squared_error',
        error_score='raise'
        )
    
    score.append(model_cv)
    mean_values.append(model_cv.mean())
    std_values.append(model_cv.std())
```

```python
# Show cross-validation result

train_score = []

df_cv = pd.DataFrame({
    'model':['dt', 'rf', 'ada', 'gbr', 'xgb', 'lgbm'],
    'mean':mean_values,
    'std':std_values
}).sort_values(by='mean', ascending=False).reset_index(drop=True)

chosen_model = df_cv.head(2)['model'].values
train_score = df_cv.head(2)['mean'].values

df_cv
```

```python
xgb = XGBRegressor(random_state=42, verbosity=0)

estimator_xgb = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=1)

random_xgb = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 120,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error'], 
    n_jobs = -1,
    refit = 'neg_root_mean_squared_error',
    random_state = 1  
)
```

```python
# randomized search
random_xgb.fit(X_train, y_train)
```

```python
# Melihat hasil tuning dalam bentuk dataframe. Diurutkan berdasarkan RMSE
pd.DataFrame(random_xgb.cv_results_).sort_values(by='rank_test_neg_root_mean_squared_error').head()['mean_test_neg_root_mean_squared_error']
```

Test the Tuned Model

```python
# Model XGBoost
model = {'XGB': XGBRegressor(random_state=1)}

# Define model terhadap estimator terbaik
xgb_tuning = random_xgb.best_estimator_

# Fitting model
xgb_tuning.fit(X_train, y_train)

# Predict test set
y_pred_xgb_tuning = xgb_tuning.predict(X_val)

# Simpan nilai metrics RMSE, MAE & MAPE setelah tuning
rmse_xgb_tuning = np.sqrt(mean_squared_error(y_val, y_pred_xgb_tuning))

score_after_tuning = pd.DataFrame({'RMSE': rmse_xgb_tuning}, index=model.keys())
score_after_tuning
```
